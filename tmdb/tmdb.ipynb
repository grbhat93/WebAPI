```python
import pandas as pd
import numpy as np
import requests
```

```python
response=requests.get("https://api.themoviedb.org/3/movie/top_rated?api_key=d02536aefd2368795dee8bc919fdd1fe&language=en-US&page=1")
```

```python
response.json()['results']
```

```python
df=pd.DataFrame(response.json()["results"])[['id','original_language','original_title','overview', 'popularity','release_date', 'title','vote_average', 'vote_count']]
df
```

```python
df.columns
```

```python
df1=pd.DataFrame()
df1
```

```python
for i in range(1,100):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=d02536aefd2368795dee8bc919fdd1fe&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','original_language','original_title','overview', 'popularity','release_date', 'title','vote_average', 'vote_count']]
    df1=df1.append(temp_df,ignore_index=True)
```
